```python
def label_encoder(df, columns):
    Encoder = dict()

    for var in columns:
        Encoder[var] = LabelEncoder().fit(df[var].astype(str))
        
    return Encoder

def encode_columns(df, columns, encoder = None):
    if not encoder:
        encoder = label_encoder(df, columns)
 
    for var in columns: 
        df[var] = encoder[var].transform(df[var].astype(str))
        
    return encoder

```

```python
def label_bin(df, columns):
    Encoder = dict()

    for var in columns:
        Encoder[var] = LabelBinarizer().fit(df[var].astype(str))
        
    return Encoder

def bin_columns(df, columns, encoder = None):
    if not encoder:
        encoder = label_bin(df, columns)
    
    for var in columns: 
        df[var] = encoder[var].transform(df[var].astype(str))

    return encoder

```
